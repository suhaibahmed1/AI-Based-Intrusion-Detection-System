```python
import os
import pandas as pd
import numpy as np

# Display all columns and some rows
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)

print("Environment ready ✅")
```

```python
DATA_PATH = "../data/raw"  # Make sure your 8 CSV files are here

csv_files = [f for f in os.listdir(DATA_PATH) if f.endswith(".csv")]
print("Found CSV files:")
print(csv_files)
```

```python
# Create a list to hold each CSV as a DataFrame
dataframes = []

for file in csv_files:
    file_path = os.path.join(DATA_PATH, file)
    df = pd.read_csv(file_path)
    dataframes.append(df)

# Merge all into one DataFrame
combined_df = pd.concat(dataframes, ignore_index=True)

print("Merged DataFrame shape:", combined_df.shape)
```

```python
# Check column names
print("Columns in dataset:")
print(combined_df.columns)

# Check label distribution (CICIDS 2017 label column has a leading space)
print("\nLabel counts:")
print(combined_df[" Label"].value_counts())
```

```python
combined_df.to_csv("../data/merged_cicids2017.csv", index=False)
print("Merged dataset saved ✅")
```
