```python
import pandas as pd
import numpy as np

# Display settings
pd.set_option("display.max_columns", None)
```

```python
merged_path = "../data/merged_cicids2017.csv"
df = pd.read_csv(merged_path)

print("Dataset loaded. Shape:", df.shape)
```

```python
# Check missing values
print("Missing values per column:")
print(df.isnull().sum().sort_values(ascending=False).head(10))

# Check for infinite values
# Select only numeric columns
numeric_cols = df.select_dtypes(include=[np.number])

# Check for infinite values in numeric columns
print("Any infinite values in numeric columns?", np.isinf(numeric_cols.values).any())
```

```python
# Replace infinite with NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Drop rows with NaN (safe for ML)
df.dropna(inplace=True)

print("After cleaning, shape:", df.shape)
```

```python
df.columns = df.columns.str.strip()
print("Columns after cleaning:")
print(df.columns)
```

```python
# Check unique labels
print("Unique labels:", df["Label"].unique())

# Optionally, encode label: BENIGN=0, ATTACK=1
df["LabelEncoded"] = df["Label"].apply(lambda x: 0 if x=="BENIGN" else 1)
print(df[["Label", "LabelEncoded"]].head())
```

```python
df.to_csv("../data/cleaned_cicids2017.csv", index=False)
print("Cleaned dataset saved ✅")
```
